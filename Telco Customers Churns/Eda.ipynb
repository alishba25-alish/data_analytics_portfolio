```python
import pandas as pd

customers_churn = pd.read_csv('C:/Users/DELL/Documents/Telco Customers Churns/WA_Fn-UseC_-Telco-Customer-Churn (1).csv')

print(customers_churn.head(5))
```

```python
print(customers_churn.info())
```

```python
print(customers_churn.isnull().sum())
```

```python
customers_churn['TotalCharges'] = pd.to_numeric(customers_churn['TotalCharges'], errors='coerce')

customers_churn.dropna(inplace=True)

# customers_churn.drop('customerID', axis=1, inplace=True)

print(customers_churn.info())
```

```python
customers_churn['SeniorCitizen'] = customers_churn['SeniorCitizen'].map({0: 'Not Senior Citizen', 1: 'Senior Citizen'})
```

```python
# insights
# Total Customer Churners 
Total_Churned = customers_churn['Churn'].value_counts()
print(Total_Churned)
```

```python
# Demographics

# By gender
Churned_by_gender = customers_churn.groupby(['Churn','gender']).size().unstack()
print(Churned_by_gender)
print()

# By SeniorCitizen
Churned_by_SeniorCitizen = customers_churn.groupby(['Churn','SeniorCitizen']).size()
print(Churned_by_SeniorCitizen)
print()

# By Partners
Churned_by_Partners = customers_churn.groupby(['Churn','Partner']).size()
print(Churned_by_Partners)
print()

# By Dependents
Churned_by_Dependents = customers_churn.groupby(['Churn','Dependents']).size()
print(Churned_by_Dependents)
print()

# By Tenure
Bins = [0,12,36,72]
labels = ['New (0-12m)', 'Mid (13-36m)', 'Loyal (37-72m)']
customers_churn['TenureGroup'] = pd.cut(customers_churn['tenure'], bins=Bins, labels=labels, right=True)

Churned_by_tenure = customers_churn.groupby(['TenureGroup','Churn']).size().unstack(fill_value=0)
print(Churned_by_tenure)
```

```python
# Customer Services
# By PhoneService
Churned_by_PhoneService = customers_churn.groupby(['Churn','PhoneService']).size()
print(Churned_by_PhoneService)
print()

# By MultipleLines
Churned_by_MultipleLines = customers_churn.groupby(['Churn','MultipleLines']).size()
print(Churned_by_MultipleLines)
print()

# By InternetService
Churned_by_InternetService = customers_churn.groupby(['Churn','InternetService']).size()
print(Churned_by_InternetService)
print()

# By OnlineSecurity
Churned_by_OnlineSecurity = customers_churn.groupby(['Churn','OnlineSecurity']).size()
print(Churned_by_OnlineSecurity)
print()

# By OnlineBackup
Churned_by_OnlineBackup = customers_churn.groupby(['Churn','OnlineBackup']).size()
print(Churned_by_OnlineBackup)
print()

# By DeviceProtection
Churned_by_DeviceProtection = customers_churn.groupby(['Churn','DeviceProtection']).size()
print(Churned_by_DeviceProtection)
print()

# By TechSupport
Churned_by_TechSupport = customers_churn.groupby(['Churn','TechSupport']).size()
print(Churned_by_TechSupport)
print()

# By StreamingTV
Churned_by_StreamingTV = customers_churn.groupby(['Churn','StreamingTV']).size()
print(Churned_by_StreamingTV)
print()

# By StreamingMovie
Churned_by_StreamingMovies = customers_churn.groupby(['Churn','StreamingMovies']).size()
print(Churned_by_StreamingMovies)
print()
```

```python
# By Contract
Churned_by_Contract = customers_churn.groupby(['Churn','Contract']).size()
print(Churned_by_Contract)
print()

# By PaperlessBilling
Churned_by_PaperlessBilling = customers_churn.groupby(['Churn','PaperlessBilling']).size()
print(Churned_by_PaperlessBilling)
print()

# By PaymentMethod
Churned_by_PaymentMethod = customers_churn.groupby(['Churn','PaymentMethod']).size()
print(Churned_by_PaymentMethod)
print()

# Average. Monthly Charges & Total Charges
print("Avg. Monthly Charges: ",round(customers_churn['MonthlyCharges'].mean(), 2))
print()
print("Avg. Total Charges: ",round(customers_churn['TotalCharges'].mean(),2))
```
